```python
# 지도에 한라산 마커 찍기
m = folium.Map(
    location = [33.36362781630753, 126.52916657111979],
    tiles = 'Stamen Terrain', # 지형을 바꿔주는 코드
    zoom_start = 10
)

tooltip = 'Click!!'

folium.Marker(
    [33.36362781630753, 126.52916657111979],
    popup = '<strong>hello world</strong>',
    tooltip = tooltip
).add_to(m)

m
```

```python
# 마커의 종류를 바꿔보자
m = folium.Map(
    location = [33.36362781630753, 126.52916657111979],
#     tiles = 'Stamen Terrain', # 지형을 바꿔주는 코드
    zoom_start = 10
)

tooltip = 'Click!!'

folium.Marker(
    [33.36362781630753, 126.52916657111979],
    popup = '<strong>한라산</strong>',
    tooltip = tooltip,
    icon = folium.Icon(color='#ff0000',icon='info-mark')
).add_to(m)

folium.Marker(
    [33.51049098299758, 126.49135573869128],
    popup = '<strong>제주 국제 공항</strong>',
    tooltip = tooltip,
    icon = folium.Icon(color='#ffffff',icon='bookmark')
).add_to(m)

folium.Marker(
    [33.50504836804892, 126.95430546321592],
    popup = '<strong>우도</strong>',
    tooltip = tooltip,
    icon = folium.Icon(color='#ffffff')
).add_to(m)

m
```

```python
m = folium.Map(
    location = [33.36362781630753, 126.52916657111979],
#     tiles = 'Stamen Terrain', # 지형을 바꿔주는 코드
    zoom_start = 10
)

tooltip = 'Click!!'

for i in range(Oreum.shape[0]):
    folium.Marker(
    [Oreum.iloc[i]['위도'], Oreum.iloc[i]['경도']],
    popup = f'<strong>{Oreum.iloc[i]["오름명"]}</strong>',
        # f string 문법을 사용, 작은 따옴표와 큰 따옴표를 위치에 맞게 구분해 사용
    tooltip = tooltip
).add_to(m)


m
```

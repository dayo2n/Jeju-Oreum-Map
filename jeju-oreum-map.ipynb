```python
# 직선 거리를 표시하고 구하는 방법

m = folium.Map(
    location = [33.36362781630753, 126.52916657111979],
    zoom_start = 10
)

tooltip = 'Click!!'

for i in range(Oreum.shape[0]):
    folium.Marker(
        [Oreum.iloc[i]['위도'], Oreum.iloc[i]['경도']],
        popup = f'<div style="width:100px"><strong>{Oreum.iloc[i]["오름명"]}</strong><br>\
            주차장 : {Oreum.iloc[i]["주차장"]}<br>\
            화장실 : {Oreum.iloc[i]["화장실"]}<br>',
        tooltip = tooltip
    ).add_to(m)
    
    folium.Circle(
        [Oreum.iloc[i]['위도'], Oreum.iloc[i]['경도']],
        radius = Oreum.iloc[i]['표고(m)'],
        color = 'red',
        fill = True
    ).add_to(m)
    
    folium.PolyLine(
        locations = [[33.51063709962625, 126.49139632094533], [Oreum.iloc[i]['위도'], Oreum.iloc[i]['경도']]],
        tooltip = tooltip
    ).add_to(m)

m
```

```python
from haversine import haversine

#haversine(위치1, 위치2)
```

```python
# 툴팁에 커서를 올렸을 때만 직선 거리를 띄워보자

m = folium.Map(
    location = [33.36362781630753, 126.52916657111979],
    zoom_start = 10
)

tooltip = 'Click!!'

for i in range(Oreum.shape[0]):
    folium.Marker(
        [Oreum.iloc[i]['위도'], Oreum.iloc[i]['경도']],
        popup = f'<div style="width:100px"><strong>{Oreum.iloc[i]["오름명"]}</strong><br>\
            주차장 : {Oreum.iloc[i]["주차장"]}<br>\
            화장실 : {Oreum.iloc[i]["화장실"]}<br>',
        tooltip = tooltip
    ).add_to(m)
    
    folium.Circle(
        [Oreum.iloc[i]['위도'], Oreum.iloc[i]['경도']],
        radius = Oreum.iloc[i]['표고(m)'],
        color = 'red',
        fill = True
    ).add_to(m)
    
    location1 = [33.51063709962625, 126.49139632094533]
    location2 = [Oreum.iloc[i]['위도'], Oreum.iloc[i]['경도']]
    distance = haversine(location1, location2, unit = 'km')
    
    folium.PolyLine(
        locations = [location1, location2],
        tooltip = f'제주국제공항에서 {Oreum.iloc[i]["오름명"]}까지 {distance}k'
    ).add_to(m)

m
```
